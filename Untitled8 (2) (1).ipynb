```python
from llama_index.core import Settings
Settings.llm = NVIDIA(model="meta/llama3-70b-instruct", base_url="https://integrate.api.nvidia.com/v1",api_key=api_key)
Settings.chunk_size = 512
from llama_index.core.memory import ChatMemoryBuffer

memory = ChatMemoryBuffer.from_defaults(token_limit=8000)
query_engine = index.as_chat_engine(
    chat_mode='context',
    memory = memory,
    system_prompt=system_prompt,
    similarity_top_k=4
)
import re

def predict(input_text):
      response = query_engine.chat(input_text)

    # Ensure the response is a string
      if not isinstance(response, str):
          response = str(response)

    # Remove the word "assistant" from the beginning of the response text
      response_text = re.sub(r'^assistant\s*', '', response)

    # Return response text
      return response_text



def get_first_five_words(text):
    words = text.split()[:5]
    return ' '.join(words)
```

```python

from flask import Flask, render_template, request, jsonify, flash
import speech_recognition as sr
from transformers import pipeline
import whisper
import os
import base64
from flask_lt import run_with_lt

from melo.api import TTS

app = Flask(__name__)
run_with_lt(app)
app.secret_key = "voice"

os.environ["CUDA_LAUNCH_BLOCKING"] = "1"



pipe = pipeline("automatic-speech-recognition",
                "openai/whisper-large-v2",
                device="cuda:0")
     


speed = 1.0
device = 'auto'  # Will automatically use GPU if available
tts_model = TTS(language='EN', device=device)
speaker_ids = tts_model.hps.data.spk2id
output_path = 'output_audio.wav'

# Chatbot routes
first_message = True

@app.route('/')
def home():
    global first_message
    first_message = True
    memory.reset()
    return render_template('chat.html')

@app.route('/predict', methods=['POST'])
def get_prediction():
    global first_message
    data = request.get_json()
    input_text = data['input']

    # Replace with your prediction logic
    prediction = predict(input_text)

    if first_message:
        first_five_words = get_first_five_words(input_text)
        first_message = False
    else:
        first_five_words = ""

    return jsonify({'response': prediction, 'first_five_words': first_five_words})

# Audio routes
@app.route('/audio_to_text/')
def audio_to_text():
    flash("Press Start to start recording audio and press Stop to end recording audio")
    memory.reset()
    return render_template('audiocall.html')

@app.route('/audio', methods=['POST'])
def audio():
    file = request.files['file']
    file.save('upload/audio.wav')

    try:
        # Load audio file into Whisper

        results = pipe("upload/audio.wav",
               chunk_length_s=30,
               batch_size=8)
        return_text = results['text']

    except sr.UnknownValueError:
        return_text = "Sorry, the audio was not clear enough to be understood."
    except Exception as e:
        return_text = f"An error occurred: {e}"

    # Pass the returned text through the predict() function
    print(return_text)
    response_text = predict(return_text)

    # Convert the response text to speech using Melo TTS


    
    # Using American accent for TTS
    tts_model.tts_to_file(response_text, speaker_ids['EN-AU'], output_path, speed=speed)
    
    with open(output_path, "rb") as audio_file:
        audio_data = base64.b64encode(audio_file.read()).decode('utf-8')

    return jsonify({'audio': audio_data, 'text': response_text})

if __name__ == "__main__":
    app.run(debug=False)
```

```python
import nest_asyncio
from fastapi import FastAPI, Request, File, UploadFile, Form
from fastapi.responses import HTMLResponse, JSONResponse
from fastapi.templating import Jinja2Templates
import whisper
import os
import base64
import boto3
import uvicorn

nest_asyncio.apply()

app = FastAPI()

os.environ["CUDA_LAUNCH_BLOCKING"] = "1"

pipe = pipeline("automatic-speech-recognition",
                "openai/whisper-large-v2",
                device="cuda:0")
     

speed = 1.0
output_path = 'output_audio2.mp3'

# Initialize Amazon Polly
polly_client = boto3.client('polly', region_name='ap-south-1',
                            aws_access_key_id='access-key',
                            aws_secret_access_key='secret-key')

# Chatbot routes
first_message = True
templates = Jinja2Templates(directory="templates")

@app.get("/", response_class=HTMLResponse)
async def home(request: Request):
    global first_message
    first_message = True
    memory.reset()  # Assuming you have a memory reset function
    return templates.TemplateResponse("chat.html", {"request": request})

@app.post("/predict", response_class=JSONResponse)
async def get_prediction(data: dict):
    global first_message
    input_text = data['input']

    # Replace with your prediction logic
    prediction = predict(input_text)

    if first_message:
        first_five_words = get_first_five_words(input_text)
        first_message = False
    else:
        first_five_words = ""

    return {"response": prediction, "first_five_words": first_five_words}

@app.get("/audio_to_text", response_class=HTMLResponse)
async def audio_to_text(request: Request):
    memory.reset()  # Assuming you have a memory reset function
    return templates.TemplateResponse("audiocall.html", {"request": request})

@app.post("/audio", response_class=JSONResponse)
async def audio(file: UploadFile = File(...)):
    file_location = f"upload/{file.filename}"
    with open(file_location, "wb") as buffer:
        buffer.write(file.file.read())

    try:
        # Transcribe audio file using Whisper
        results = pipe("upload/audio.wav",
               chunk_length_s=30,
               batch_size=8,)
        return_text = results['text']

    except Exception as e:
        return_text = f"An error occurred: {e}"

    # Pass the returned text through the predict() function
    print(return_text)
    response_text = predict(return_text)

    # Convert the response text to speech using Amazon Polly
    response = polly_client.synthesize_speech(
        Engine='neural',
        Text=response_text,
        OutputFormat='mp3',
        VoiceId='Kajal'  # Change the VoiceId as needed
    )

    # Save the audio stream returned by Amazon Polly
    with open(output_path, 'wb') as audio_file:
        audio_file.write(response['AudioStream'].read())

    with open(output_path, "rb") as audio_file:
        audio_data = base64.b64encode(audio_file.read()).decode('utf-8')

    return {"audio": audio_data, "text": response_text}

if __name__ == "__main__":
    uvicorn.run(app, host="127.0.0.1", port=8000)
```
